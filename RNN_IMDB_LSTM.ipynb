# RNN IMDb LSTM (Minimal)
Minimal, clean notebook for sentiment analysis using LSTM.

```
import tensorflow as tf
from tensorflow.keras import layers, models, preprocessing

print("TF version:", tf.__version__)
```

```
vocab = 10000
maxlen = 200

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.imdb.load_data(num_words=vocab)
x_train = preprocessing.sequence.pad_sequences(x_train, maxlen=maxlen)
x_test = preprocessing.sequence.pad_sequences(x_test, maxlen=maxlen)

x_train.shape, y_train.shape
```

```
model = models.Sequential([
    layers.Embedding(vocab, 128, input_length=maxlen),
    layers.LSTM(128),
    layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```
history = model.fit(x_train, y_train, epochs=3, batch_size=128, validation_split=0.2)
```

```
model.evaluate(x_test, y_test)
```
